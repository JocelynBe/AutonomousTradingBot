```python
from typing import List, Union

class Solution:


    def normalize(self, s: str) -> List[Union[int, str]]:
        pos = 0
        n = len(s)
        ans = []
        prev_is_numeric = False
        while pos < n:
            c = s[pos]
            if c == ' ':
                pos += 1
                continue
            elif prev_is_numeric and c.isnumeric():
                ans[-1] = ans[-1] + c
            elif c.isnumeric():
                ans.append(c)
            elif ans:
                ans[-1] = int(ans[-1])
                ans.append(c)
            else:
                assert False

            prev_is_numeric = c.isnumeric()
            pos += 1
        if ans[-1].isnumeric():
            ans[-1] = int(ans[-1])
        return ans

    def calculate(self, s: str) -> int:
        l = self.normalize(s)
        print(l)
        stack = []
        pos = 0
        n = len(l)
        while pos < n:
            c = l[pos]
            if isinstance(c, int):
                stack.append(c)
                pos += 1
            elif c == '*':
                val = stack.pop()
                stack.append(val*l[pos+1])
                pos += 2
            elif c == '/':
                val = stack.pop()
                stack.append(int(val/l[pos+1]))
                pos += 2
            elif c == '+':
                stack.append('+')
                pos += 1
            elif c == '-':
                stack.append('-')
                pos += 1
            else:
                assert False, c

        print(stack)

        res = 0
        sgn = 1
        for c in stack:
            print(res)
            if isinstance(c, int):
                res += sgn * c
            elif c == '+':
                sgn = 1
            elif c == '-':
                sgn = -1
            else:
                assert False

        return res
```
